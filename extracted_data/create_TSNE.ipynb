```python
import pandas as pd
from sklearn.manifold import TSNE

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def TSNE_visualization(round_soc_list):
    # TSNE visualization with standardScaler data vectors
    principalComponents_standardScaler_with_SOC_code = pd.read_csv('./occupations_as_columns_with_SOC_CODE.csv').set_index('Unnamed: 0').T

    sub_standardScaler_with_SOC = principalComponents_standardScaler_with_SOC_code[principalComponents_standardScaler_with_SOC_code["round O*NET-SOC Code"].isin(round_soc_list)]
    sub_tandardScaler_without_SOC = sub_standardScaler_with_SOC.drop(["round O*NET-SOC Code","O*NET-SOC Code"],axis =1)

    standardScaler_TSNE=TSNE(learning_rate = 25, perplexity=5, n_iter=1000)
    standardScaler_TSNE_features = standardScaler_TSNE.fit_transform(sub_tandardScaler_without_SOC)

    sub_standardScaler_with_SOC['TSNE vector 1'] = standardScaler_TSNE_features[:,0]
    sub_standardScaler_with_SOC['TSNE vector 2'] = standardScaler_TSNE_features[:,1]

    sns.scatterplot(
        x = 'TSNE vector 1', y = 'TSNE vector 2',
        hue="round O*NET-SOC Code",
        palette=sns.color_palette("hls",5),
        data = sub_standardScaler_with_SOC,
        legend="full",
        alpha=1
    )
    plt.show
```

```python
# get subset of occupation by round SOC code
# [113031.0, 119121.0, 131081.0, 131199.0, 172051.0, 172141.0, 191031.0, 292011.0, 292099.0, 333021.0, 518013.0, 536051.0]
round_SOC = [113031.0,131199.0,172051.0,172141.0,  292099.0 ]
TSNE_visualization(round_SOC) # familiar occupations
```

```python
round_SOC = [119121.0, 131081.0, 191031.0, 292011.0, 518013.0]
TSNE_visualization(round_SOC) # unfamiliar occupations
```
